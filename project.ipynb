````python
For Loop:
الفور لووب يستخدم لتكرار مجموعة من الأوامر على مجموعة من العناصر. يمكن أن تكون هذه العناصر قائمة، سلسلة، ترتيب أو هياكل بيانات أخرى.1. تكرار على عناصر قائمة:


# قائمة من الفواكه
fruits = ["تفاحة", "موز", "فراولة", "عنب"]

# فور لووب لطباعة كل فاكهة
for fruit in fruits:
    print(f"أنا أحب {fruit}")

# فور لووب لطباعة أعداد من 1 إلى 5
for i in range(1, 6):
    print(i)
While Loop:
الوايل لووب يستخدم لتكرار مجموعة من الأوامر حتى تصبح شرط معينة غير صحيحة.1. تكرار حتى تصبح متغير معينة أكبر من قيمة محددة:

# وايل لووب لطباعة أعداد من 1 إلى 5
counter = 1
while counter <= 5:
    print(counter)
    counter += 1

# وايل لووب لطباعة أعداد زوجية حتى تصل إلى 10
num = 2
while num <= 10:
    print(num)
    num += 2

# وايل لووب حتى يكون العدد زوجيًا وأقل من 10
num = 2
while num % 2 == 0 and num < 10:
    print(num)
    num += 2
Material
Abdul Omair posted a new material: Functions
Posted 2 Feb
الدوال في بايثون:


الدوال هي كتل من الكود يمكن تنفيذها
عند الحاجة، وهي تسمح بتنظيم البرنامج وإعادة استخدام الكود. تعرف الدوال باستخدام
الكلمة الرئيسية def،
وتأخذ مجموعة من البيانات (المعاملات) كإدخال وتقوم بإرجاع قيمة (أو قيم) في بعض
الأحيان.





def greet(name):


    """تحية بسيطة"""


    print(f"مرحبًا، {name}!")





# استدعاء الدالة


greet("أحمد")











في هذا المثال:





greet
هو اسم الدالة.


(name)
هو المعامل الذي تستخدمه الدالة.


"""تحية
بسيطة""" هو السطر الوثائقي الذي يوضح وظيفة الدالة.


لتعريف دالة، نستخدم الكلمة الرئيسية def، ونحدد اسم الدالة والمعاملات بين الأقواس،
ونستخدم نقطتين لتحديد بداية الكود الخاص بالدالة.





قيم العودة (Return
Values):


يمكن للدوال في بايثون إرجاع قيم
باستخدام كلمة return.
على سبيل المثال:





def add_numbers(x, y):


    """جمع اثنين من الأرقام"""


    result = x + y


    return result





# استدعاء الدالة


sum_result = add_numbers(5, 3)


print(f"المجموع: {sum_result}")








المعاملات الافتراضية:


يمكن تحديد قيم افتراضية للمعاملات في
الدوال. هذا يعني أنه يمكن استدعاء الدالة بدون توفير قيم لتلك المعاملات.





def greet(name="زائر"):


    """تحية بسيطة"""


    print(f"مرحبًا، {name}!")





# استدعاء الدالة بمعامل


greet("أحمد")





# استدعاء الدالة بدون معامل


greet()











المعاملات المتغيرة:


يمكن لدوال بايثون أن تأخذ عددًا
متغيرًا من المعاملات باستخدام *args
و **kwargs.





def print_args(*args, **kwargs):


    """طباعة المعاملات المتغيرة"""


    print("المعاملات:")


    for arg in args:


        print(arg)





    print("المعاملات الإضافية:")


    for key, value in kwargs.items():


        print(f"{key}: {value}")





# استدعاء الدالة مع معاملات متغيرة


print_args(1, "hello", name="John", age=25)








هذا هو نظرة عامة عن الدوال في بايثون. تتيح
الدوال تنظيم الكود وإعادة استخدامه بشكل فعال، وهي جزء أساسي من بنية أي برنامج.
Material
Abdul Omair posted a new material: Data Types
Posted 2 Feb
في بايثون هناك العديد من أنواع البيانات المدمجة
المهمة التي يمكن استخدامها لتمثيل المعلومات. إليك بعض الأنواع الأساسية:





1.
int (الأعداد
الصحيحة): تُستخدم لتمثيل الأعداد الصحيحة، مثل 1 و -5 و
100.





    ```python


    x = 5


    ```





2.
float (الأعداد
العائمة): تُستخدم لتمثيل الأعداد العائمة (الكسور أو الأعداد العشرية).





    ```python


    y = 3.14


    ```





3.
str (النصوص):
تُستخدم لتمثيل النصوص، ويمكن تحديدها بين علامتي اقتباس (فردية أو مزدوجة).





    ```python


    name = "John"


    ```





4.
bool (القيم
البولية): تُستخدم لتمثيل القيم البولية (True أو False).





    ```python


    is_student = True


    ```





````
